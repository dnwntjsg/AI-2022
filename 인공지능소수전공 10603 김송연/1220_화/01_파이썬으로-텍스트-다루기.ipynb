```python
print(kkma.pos('질문이나 건의사항은 깃헙 이슈 트래커에 남겨주세요.'))
```

```python
out = kkma.pos('질문이나 건의사항은 깃헙 이슈 트래커에 남겨주세요.')

for word, tag in out:
  if tag[0] in ('N', "V"):
    print(word, tag)
```

# 정규 표현식

```python
import re
```

```python
exp = r"\+?[0-9][0-9\-]{7,14}"
text = "안녕하세요 제 이름은 최이준이고, 연락처는 010-9875-8572 입니다."

print(re.findall(exp, text))

print(re.sub(exp, "<phone>", text))
```

```python
exp = r"([\w\.]+)@gmail\.com"
text = "나의 이메일 주소는 haru060806@gmail.com 입니다."

print(re.sub(exp, r"지메일(\1)", text))
```

```python
exp = "(\n)+"
text = """
투명 드래곤이 울부짖었다

우와아앙

우앙






구아아앙
"""

print(re.sub(exp, r"\1", text))
```

```python
exp = r"[^\w\d\s가-힇ㅋ-ㅎ]+"
def preprocess(text):
  text = emoticon_normalize(text)
  text = re.sub(exp, "", text)
  return text

data["preprocessed_text"] = data.document.map(preprocess)
data["preprocessed_length"] = data.preprocessed_text.str.len()
data
```
