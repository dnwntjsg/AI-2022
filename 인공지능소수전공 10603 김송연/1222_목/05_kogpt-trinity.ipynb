```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation", model="skt/ko-gpt-trinity-1.2B-v0.5", device='cuda:0')
```

```python
generator(
    "나는 무엇인가 ",
    max_length=256,
    do_sample=True
)
```

```python
def generate(prompt, **kwargs):
  out=generator(
      prompt,
      **kwargs
  )[0]['generated_text']

  print(out)

def generate_s(prompt, tokens=32):
  generate(
      prompt,
      do_sample=True,
      top_p=0.7,
      max_new_tokens=tokens,
      repetition_penalty=1.2
  )

def generate_g(prefix, tokens=32):
  generate(prefix, max_new_tokens=tokens)
```

```python
generate_g("""
문장을 읽고 긍증/부정 중 하나를 선택하세요
아 이 영화 진짜 별로다=부정
올해 최고의 망작=부정
정말 재미있었어요=긍정
최고의 영화였습니다=긍정
엄복동 급의 영화였다=""")
```

```python
generate_s("""제시된 단어를 보고 대통령을 욕하세요
탄핵=ㅉㅉ뽑아줬더니 뻘짓만 하네 탄핵하자
하야=저 미친놈 알아서 하야해라
사퇴=""")
```
