```python
from transformers import pipeline
```

```python
model = pipeline(
    'text2text-generation',
    model='yuchenlin/BART0'
)
```

```python
model(
    """News title: Kim Jong-un launches nuclear missile
Look at the headline above and make a news headline
content: """,
    max_length=256,
    min_length=128
)
```

```python
model("""
Question: Who is Donald Trump?
Answer the above question""")
```
